```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Lendo arquivo com metadados

```python
musicnet_path = "musicnet_metadata.csv"
musicnet_metadata = pd.read_csv(musicnet_path)
musicnet_metadata.head()
```

```python
musicnet_metadata.info()
```

```python
musicnet_metadata.nunique()
```

## Distribuição por artista

```python
# Otimizar com Dict Compreheensions
composer = musicnet_metadata["composer"].value_counts()

label = [composer.index.tolist()]
plt.pie(composer, shadow=True, autopct='%1.1f%%', startangle=90)
plt.gcf().set_size_inches(12,12)
plt.legend(labels=label[0], loc="best")
plt.show()
```

## Distribuição por "ensembles"

```python
# Otimizar com Dict Compreheensions
ensemble = musicnet_metadata["ensemble"].value_counts()

label = [ensemble.index.tolist()]
plt.pie(ensemble, shadow=True, autopct='%1.1f%%', startangle=90)
plt.gcf().set_size_inches(12,12)
plt.legend(labels=label[0], loc="best")
plt.show()
```
